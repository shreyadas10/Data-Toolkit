2. How does broadcasting work in NumPy ?

-> Broadcasting is a powerful feature in NumPy that allows you to perform operations on arrays with different shapes and sizes. Here's how it works:

-> Rules of Broadcasting:-
 - Matching dimensions: When operating on two arrays, NumPy compares their shapes element-wise. It starts with the trailing dimensions, and works its way forward.

 -  Singleton dimensions: If one array has a singleton dimension (i.e., a dimension of size 1), NumPy will broadcast that dimension to match the corresponding dimension of the other array.

 -  Size 1 arrays: If one array has a size of 1, it can be broadcasted to any shape.

6. What are the differences between NumPy arrays and Python lists ?

->NumPy arrays and Python lists are both data structures used to store collections of data, but they have distinct differences in terms of their characteristics, advantages, and use cases:-

 -> Differences:

-  Homogeneous vs. Heterogeneous: NumPy arrays are homogeneous, meaning all elements must be of the same data type. Python lists, on the other hand, are heterogeneous, allowing elements of different data types.
- Memory Layout: NumPy arrays store data in a contiguous block of memory, which allows for efficient caching and vectorized operations. Python lists, by contrast, store data as a collection of pointers to individual elements, which can lead to slower performance.
- Performance: NumPy arrays are generally faster than Python lists for numerical computations, thanks to their contiguous memory layout and optimized C code. Python lists, however, can be faster for certain operations like insertion and deletion.
-Multi-Dimensionality: NumPy arrays can be multi-dimensional, allowing for efficient representation of matrices, tensors, and other high-dimensional data structures. Python lists can also be nested, but they don't support the same level of multi-dimensionality as NumPy arrays.
- Indexing and Slicing: Both NumPy arrays and Python lists support indexing and slicing, but NumPy arrays provide more advanced indexing and slicing capabilities.

21.  Why is NumPy’s array slicing faster than Python’s list slicing ?

-> NumPy's array slicing is faster than Python's list slicing for several reasons:

- Vectorized operations: NumPy arrays are designed to support vectorized operations, which means that operations are applied to the entire array at once, rather than iterating over individual elements. This allows NumPy to take advantage of low-level optimizations and parallelization.
-Memory layout: NumPy arrays are stored in a contiguous block of memory, which allows for efficient caching and memory access. This is in contrast to Python lists, which are stored as a collection of pointers to individual objects, leading to more memory accesses and slower performance.
- Strided memory access: NumPy arrays use a strided memory access pattern, which means that the memory address of each element is calculated using a simple offset from the base address. This allows NumPy to access elements in a predictable and efficient manner.

4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap ?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

corr_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

13. Create a 3D scatter plot using Plotly.

```python
import plotly.graph_objects as go

np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)


df = pd.DataFrame({'x': x, 'y': y, 'z': z})

fig = go.Figure(data=[go.Scatter3d(
    x=df['x'],
    y=df['y'],
    z=df['z'],
    mode='markers',
    marker=dict(
        size=5,
        color=z,
        colorscale='Viridis',
        opacity=0.8
    )
)])


fig.update_layout(
    title='3D Scatter Plot',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ),
    width=700,
    height=700
)

fig.show()
```
